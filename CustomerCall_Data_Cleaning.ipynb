```python
import pandas as pd
```

```python
df = pd.read_excel(r'/content/Customer Call List.xlsx')
df
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
df["Last_Name"].str.strip()
df["Last_Name"] = df["Last_Name"].str.replace("[^a-zA-Z]","")
```

```python
#Clean and Standardize the phone numbers using replace and lambda function
df["Phone_Number"] = df["Phone_Number"].str.replace("[^0-9]","")
df["Phone_Number"] = df["Phone_Number"].apply(lambda n : str(n))
df["Phone_Number"].apply(lambda n : n[0:3] + '-' + n[3:6] + '-' + n[6:10])
df["Phone_Number"] = df["Phone_Number"].replace("nan", "")
df["Phone_Number"]
```

```python
df[["Street_Address","State", "Zip_Code"]] = df["Address"].str.split(",", 2, True)
df
```

```python
#Clean and Standardize Paying Customer Column
df["Paying Customer"] = df["Paying Customer"].replace("N/a","")
df["Paying Customer"] = df["Paying Customer"].replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].replace("No","N")
```

```python
#Clean and Standarize Do_Not_Contact Column
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("No", "N")

```

```python
#Remove the Not_Useful_Column
df.drop("Not_Useful_Column",axis = 1, inplace = True)
```

```python
df = df.replace("N/a", "")
df = df.fillna("")
df
```

```python
#Drop the rows with no contact number
for i in df.index:
  if df.loc[i, "Phone_Number"] == "":
    df.drop(i, inplace = True)
df
```

```python
#Reset the index
df.reset_index(inplace = True)
```
