```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/food_coded.csv')
df
```

```python
#Check for duplicates and remove
df.drop_duplicates()

#Check for null rows
df.isnull()
```

```python
df["GPA"] = df["GPA"].str.strip()
df["GPA"] = df["GPA"].str.replace('[^0-9.]','')
df["GPA"]
```

```python
#Change the Gender column data from 1 and 2 to F(Female) and M(Male)

df["Gender"] = df["Gender"].astype(str)
df["Gender"].str.strip()
df["Gender"] = df["Gender"].replace("2", "M")
df["Gender"] = df["Gender"].replace("1", "F")
df["Gender"]
```

```python
#Renaming the breakfast column to a more descriptive one
df.rename(columns={'breakfast': 'Breakfast_Preference'}, inplace=True)
df["Breakfast_Preference"]
```

```python

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth',None)
```

```python
#Change the datatype from float to int
#Fill the null values with -1
#Integer values refer to answers in survey
df["calories_day"] = df["calories_day"].fillna('-1').astype(int)
```

```python
#Change the datatype from float to int
#Fill the null values with -1
#Guessed calories
df["calories_scone"] = df["calories_scone"].fillna('-1').astype(int)
```

```python
#Cleaning the textual data
#Remove the ellipsis, backslashs, dashes and more
#Convert to lowercase

df["comfort_food"] = df["comfort_food"].str.strip()
df["comfort_food"] = df["comfort_food"].str.replace('[^a-zA-Z, ]','')
df["comfort_food"] = df["comfort_food"].str.replace(' and ', ' ')
df["comfort_food"] = df["comfort_food"].str.lower()
```

```python

def map_to_category(comfort_foods):
    categories = set()
    for food in comfort_foods.split(', '):
        for category, category_items in category_mapping.items():
            if food in category_items:
                categories.add(category)
                break
    return ', '.join(categories)

df['comfort_food_categorized'] = df["comfort_food"].apply(map_to_category)
df['comfort_food_categorized']
```
